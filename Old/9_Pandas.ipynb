# Traditional way of reading structure dataset => hard way

#### Method - 1

```python
path = r"data/mobile_data.csv"

data = []
with open(path,mode="r") as file:
    for line in file.readlines():
        elements = line.split(",")
        elements[-1] = elements[-1].replace("\n","")
        data.append(elements)
```

```python
import numpy as np

data = np.array(data[1:])
data = data.astype(float)
```

```python
data.mean(axis=0)
```

# Indexing, Selecting and Slicing Data in Pandas

Pandas now supports three types of Multi-axes indexing:

**.loc()**\
Pandas provide various methods to have purely **label based indexing**. When slicing, the start bound is also included. Integers are valid labels, but they refer to the label and not the position.

**.iloc()**\
Pandas provide various methods in order to get purely **integer based indexing**. Like python and numpy, these are 0-based indexing.

## loc

```python
data["sim_type"]
```

```python
data.loc[:,"sim_type"] # data.loc[row, col]
```

```python
data.loc[0:5,"sim_type"]
```

```python
data.loc[:,['sim_type', 'aperture','front_camera_resolution']]
```

```python
data.loc[100:,['sim_type', 'aperture','front_camera_resolution']]
```

```python
data.loc[100,['sim_type', 'aperture','front_camera_resolution']]
```

## iloc

```python
data.iloc[:,:]
```

```python
data.iloc[:,['sim_type', 'aperture','front_camera_resolution']] # wrong way to give
```

```python
data.iloc[:,0:3]
```

```python
data.iloc[:,4:]
```

# Condition & slicing in pandas

```python
data[data["processor_frequency"] >= 2.5]
```

```python
data[(data["processor_frequency"] >= 2.5) & (data["stand_by_time"] >= 300)]
```

```python
data[(data["processor_frequency"] >= 2.5) & (data["stand_by_time"] >= 300)][["sim_type","aperture"]] 
# -> chaining (not to use all the time; not a efficient way)
```

```python
data.loc[(data["processor_frequency"] >= 2.5) & (data["stand_by_time"] >= 300), ["sim_type","aperture"]]
```

```python
data.iloc[data['processor_frequency'] > 2.3 , 4:5] # will not work for bool condition
```

# Sample complex condition in pandas

```python
data[ (data['processor_frequency'] > 2.3) & (data["sim_type"] ==3) ]
```

```python
data.loc[ (data['processor_frequency'] > 2.3) & (data["sim_type"] ==3), ["gpu_rank","weight","stand_by_time"] ]
```
